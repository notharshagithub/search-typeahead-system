```python
import pandas as pd

# Load the parquet file into a DataFrame
df = pd.read_parquet('/content/shopping_queries_dataset_examples.parquet')

# Display the first 5 rows of the DataFrame

df
```

```python
# Drop duplicate rows based on the 'query' column
df_unique_queries = df.drop_duplicates(subset=['query']).copy()

print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame with unique queries shape: {df_unique_queries.shape}")

display(df_unique_queries.head())
```

```python
import numpy as np

# Keep only the 'query' column
df_queries_with_random_id = df_unique_queries[['query']].copy()

# Generate a new column with random integers from 1 to 100,000
df_queries_with_random_id['random_id'] = np.random.randint(1, 100001, size=len(df_queries_with_random_id))

# Display the head of the new DataFrame
display(df_queries_with_random_id.head())
```

```python
# Rename the 'random_id' column to 'count'
df_queries_with_random_id = df_queries_with_random_id.rename(columns={'random_id': 'count'})

# Display the head of the DataFrame with the renamed column
display(df_queries_with_random_id.head())
```

```python
df_typeahead = pd.read_csv('typeahead_dataset.csv')
```

```python
df_typeahead = df_typeahead.dropna(subset=['query'])
df_typeahead = df_typeahead[df_typeahead['query'].str.strip() != '']
print(f"DataFrame shape after removing null/empty queries: {df_typeahead.shape}")
```

```python
df_typeahead.to_json('typeahead_dataset.json', orient='records', indent=4)
print("Cleaned DataFrame saved as 'typeahead_dataset.json'")
```

```python
df_queries_with_random_id.to_csv('typeahead_dataset.csv', index=False)
print("DataFrame saved as 'typeahead_dataset.csv'")
```
